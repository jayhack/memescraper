```python
import os
data_dirs = os.listdir('./data/imgur')
filenames = []
for data_dir in data_dirs:
    filenames += [os.path.join('./data/imgur', data_dir, x) for x in os.listdir(os.path.join('./data/imgur', data_dir))]
filenames = filter(lambda x: x.endswith('.pkl'), filenames)
filenames
```

```python
import pickle
memes = []
for filename in filenames:
    memes += pickle.load(open(filename, 'r'))
```

```python
ids = [m.id for m in memes]
len(set(ids))
```

```python
contents = [m.content for m in memes]
```

```python
names = [c.split(':')[0] for c in contents if ':' in c]
```

```python
names = filter(lambda x: x.endswith('Meme'), names)
```

```python
from collections import Counter
name_counter = Counter(names)
```

```python
name_counter.most_common()
```
